## Reading the data

In the next cell, use Python to obtain data from the API endpoint you chose in the previous step. Workflow:

1. Import libraries.

1. Define the initial API endpoint URL.

1. Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.

1. Find out the total page count in the 1st page data.

1. Use a FOR loop to make requests for the additional pages from 2 to page count. Append the data of each additional page to the flatterned data object.

1. Print and review the data you obtained.

```python
# your code here
#1 Import libraries.
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
import urllib

#2 Define the initial API endpoint URL.
url='https://www.zalando.fr/api/catalog/articles?categories=promo-femme&limit=84&offset=84'

#3 Make request to obtain data of the 1st page. Flatten the data and store it in an empty object variable.
json_url = urllib.request.urlopen(url)
results = json.loads(json_url.read())

flattened_data = json_normalize(results)

flattened_data1 = json_normalize(flattened_data.articles[0])
flattened_data1
```

```python
#4 Find out the total page count in the 1st page data.
total_pages=results['pagination']['page_count']
total_pages
```

```python
#5 Use a FOR loop to make requests for the additional pages from 2 to page count. 
#Append the data of each additional page to the flatterned data object.
df=pd.DataFrame()
#for i in range(total_pages): no me funcionó con las páginas totales así que lo reduje para fines prácticos
for i in range(100):
    k=84*i
    url=f'https://www.zalando.fr/api/catalog/articles?categories=promo-femme&limit=84&offset={k}&sort=sale'
    response = urllib.request.urlopen(url,timeout=10)
    results = json.loads(response.read())
    flattened_data = json_normalize(results)
    flattened_data1 = json_normalize(flattened_data.articles[0])
    flattened_data1=flattened_data1.set_index('sku')
    df = df.append(flattened_data1)

#6Print and review the data you obtained.
df
```

## Bonus

Extract the following information from the data:

* The trending brand.

* The product(s) with the highest discount.

* The sum of discounts of all goods (sum_discounted_prices divided by sum_original_prices).

```python
# your code here

#The trending brand.
df.brand_name.value_counts().index[0]
```

```python
#The product(s) with the highest discount. THE TOP 10
df['price.original']=df['price.original'].str.extract('(\d*,\d*)')
df['price.promotional']=df['price.promotional'].str.extract('(\d*,\d*)')

df['price.original'] = [x.replace(',', '.') for x in df['price.original']]
df['price.promotional'] = [x.replace(',', '.') for x in df['price.promotional']]
df['discount_amount']=df['price.original'].astype(float)-df['price.promotional'].astype(float)
df1=df.copy()
total_disc=df1.groupby(['brand_name']).sum().discount_amount
```

```python
total_disc.sort_values(ascending=False).index[0:10]
```

```python
#The sum of discounts of all goods (sum_discounted_prices divided by sum_original_prices).
df1['price_original'] = df['price.original'].astype(float)
df1['price_promotional'] = df['price.promotional'].astype(float)
#df['price.promotional']
total_POrig=df1.groupby(['brand_name']).sum().price_original
total_PDisc=df1.groupby(['brand_name']).sum().price_promotional
total=df1.groupby(['brand_name']).sum().price_promotional/df1.groupby(['brand_name']).sum().price_original
total
```
